        • What does your indicator measure?
        • When does your strategy enter/exit?
        • Why might this work (behavioural explanation)?
    
    The indicator measures the Rate of Change (ROC) over a 10-day period, quantifying the percentage price movement from 10 days ago to today. The strategy enters a long position when ROC exceeds the 3% threshold, signaling strong upward momentum, and exits to cash when ROC falls below 3%. This works because of investor underreaction and herding behavior—when prices show sustained momentum above the threshold, it indicates new information is being gradually absorbed into the market, and as more investors observe the trend and react, the momentum becomes self-reinforcing. The 3% threshold acts as a noise filter by distinguishing between random daily price fluctuations (which rarely exceed 3% over 10 days) and genuine momentum trends that reflect meaningful information flow, preventing the strategy from overtrading on insignificant price movements while still capturing profitable trend-following opportunities.
