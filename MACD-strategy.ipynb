```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt

QUIET = True
%run loading-price-data.ipynb
%run backtesting-framework.ipynb

# Moving Average Convergence Divergence (MACD)

def macd_momentum_strategy(data, fast = 12, slow = 26, signal_line = 9):
    data = data.copy()
    #calculate the two emas
    fast_ema = data['Close'].ewm(span = fast, adjust = False).mean()
    slow_ema = data['Close'].ewm(span = slow, adjust = False).mean()
    #calculate the macd line (difference between fast and slow ema)
    macd_line = fast_ema - slow_ema
    signal_line_values = macd_line.ewm(span = signal_line, adjust = False).mean()
    signal = np.where(macd_line > signal_line_values, 1, 0)
    return pd.DataFrame({'MACD_signal': signal}, index = data.index)

signal_macd = macd_momentum_strategy(data, fast = 12, slow = 26, signal_line = 9)
```
