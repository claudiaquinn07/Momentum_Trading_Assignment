Moving Average Crossover Momentum Strategy

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt

QUIET = True
%run loading-price-data.ipynb
%run backtesting-framework.ipynb

#Moving Average Crossover Momentum Strategy
def ma_crossover_momentum_strategy(data, fast_n = 20, slow_n = 50):
    data = data.copy()
    fast_ma = data['Close'].rolling(window=fast_n).mean()
    slow_ma = data['Close'].rolling(window=slow_n).mean()
    signal = np.where(fast_ma>slow_ma, 1, 0)
    return pd.DataFrame({'MA_signal': signal}, index=data.index)

signal_ma = ma_crossover_momentum_strategy(data, fast_n = 20, slow_n = 50)
```
