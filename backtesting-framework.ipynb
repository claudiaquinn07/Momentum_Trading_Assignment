Backtesting Framework

```python
#Backtest Framework
def backtest_framework(data, signal):
   data = data.copy()
   data['Signal'] = signal.values
   data['Position'] = data['Signal'].shift(1)


   #calculate returns
   data['Strategy_Returns'] = data['Position'] * data['Close'].pct_change()
   data['Buy_and_Hold_Returns'] = data['Close'].pct_change()
   data.dropna(inplace=True)
   return data
```
